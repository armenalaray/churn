```python
import pandas as pd

import matplotlib.pyplot as plt

import os

churn_data = pd.read_csv('scores.csv', index_col=[0,1])
```

```python
met1 = churn_data['post_per_month']
met2 = churn_data['like_per_month']
```

```python
corr = met1.corr(met2)
```

```python
plt.scatter(met1, met2, marker='.', color='black', linewidth=1)
plt.xlabel('post_per_month')
plt.ylabel('like_per_month')
plt.title('Correlation = %.5f' % corr)
plt.tight_layout()
plt.grid()
plt.savefig('correlation.png')
plt.close()
```
