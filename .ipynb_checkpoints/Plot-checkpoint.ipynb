```python
import pandas as pd
import matplotlib.pyplot as plt

from math import ceil

csvs = [
    ("adview","socialnet7\\socialnet7_metric_stats_over_time_adview_per_month.csv"),
    ("dislike", "socialnet7\\socialnet7_metric_stats_over_time_dislike_per_month.csv"),
    ("like", "socialnet7\\socialnet7_metric_stats_over_time_like_per_month.csv"),
    ("newfriend", "socialnet7\\socialnet7_metric_stats_over_time_newfriend_per_month.csv"),
    ("post", "socialnet7\\socialnet7_metric_stats_over_time_post_per_month.csv"),
    ("reply", "socialnet7\\socialnet7_metric_stats_over_time_reply_per_month.csv"),
    ("unfriend", "socialnet7\\socialnet7_metric_stats_over_time_unfriend_per_month.csv"),
]

for metric_data_path in csvs:
    qa_data_df = pd.read_csv(metric_data_path[1])

    plt.figure(figsize=(6,6))
    
    plt.subplot(4,1,1)
    plt.plot('calc_date', 'max', data=qa_data_df, marker='', linestyle=None, color='black', linewidth=2, label='max')
    plt.ylim(0, ceil(1.1 * qa_data_df['max'].dropna().max()))
    plt.legend()
    
    plt.subplot(4,1,2)
    plt.plot('calc_date', 'avg', data=qa_data_df, marker='', linestyle=None, color='black', linewidth=2, label='avg')
    plt.ylim(0, ceil(1.1 * qa_data_df['avg'].dropna().max()))
    plt.legend()
    
    plt.subplot(4,1,3)
    plt.plot('calc_date', 'min', data=qa_data_df, marker='', linestyle=None, color='black', linewidth=2, label='min')
    plt.ylim(0, ceil(1.1 * qa_data_df['min'].dropna().max()))
    plt.legend()
    
    plt.subplot(4,1,4)
    plt.plot('calc_date', 'n_calc', data=qa_data_df, marker='', linestyle=None, color='black', linewidth=2, label='n_calc')
    plt.ylim(0, ceil(1.1 * qa_data_df['n_calc'].dropna().max()))
    plt.legend()
    
    
    plt.title(label=metric_data_path[0], y=4.7, loc='center')
    #esta es para formatear la linea
    plt.gca().figure.autofmt_xdate()

    save_to_path = metric_data_path[1].replace('.csv','.png')
    print(save_to_path)
    
    plt.savefig(save_to_path)

    plt.close()

    
```
