```python
import pandas as pd
import numpy as np
import os
```

```python
score_data = pd.read_csv('scores.csv', index_col=[0,1])
```

```python
data_2group = score_data.drop('is_churn',axis=1)
```

```python
data_2group
```

```python
load_mat_df = pd.read_csv('load_matrix.csv', index_col=[0])
```

```python
load_mat_ndarray = load_mat_df.to_numpy()
```

```python
ndarray_2group = data_2group[load_mat_df.index.values].to_numpy()
```

```python
ndarray_2group.shape
```

```python
grouped_ndarray = np.matmul(
    ndarray_2group,
    load_mat_ndarray)
```

```python
churn_data_grouped = pd.DataFrame(
    grouped_ndarray, 
    columns=load_mat_df.columns.values, 
    index=score_data.index
)
```

```python
churn_data_grouped['is_churn'] = score_data['is_churn']
churn_data_grouped
```

```python
churn_data_grouped.to_csv('groupscore.csv')
```
